```python
f = lambda x: cubic_xc_upper(x) - cubic_xc_lower(x)

# Kita gunakan Simpson's Rule 1/3 maka data harus genap
x = np.linspace(0, 1, 500)

def simpson_rule(f, x):
    diff = x[1] - x[0]
    size = x.size
    y = 0
    simpson1per3 = 0
    # Simpson 1/3 Rule
    for i in range(0, size):
        yIntegral = 0
        # Pertama dan akhir tidak dikali apa - apa.
        if i == 0 or i == size - 1:
            yIntegral = f(x[i])
        else:
            # Mempunyai pola 1 4 2 4 ... 2 4 1
            # Ketika i adalah ganjil maka dikali 2
            # Ketika i adalah genap maka dikali 4
            if i % 2 == 1:
                yIntegral = 2 * f(x[i])
            else:
                yIntegral = 4 * f(x[i])
        simpson1per3 += yIntegral
    simpson1per3 = simpson1per3 * diff / 3
    y = simpson1per3
    return y


print(f"Lift Coefficient (Cl) = {simpson_rule(f, x)}")
```
