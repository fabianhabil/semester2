```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
def my_lin_interp(x, y, X):
    # Buar Array kosong untuk menyimpan titik interpolasi X
    Y = []
    # Loop sebanyak X yang akan dicari interpolasinya
    for i in range(0, len(X)):
        # Deklarasi index untuk nanti dicari di data sampel sesuai data X yang akan dicari
        # interpolasinya.
        index = -1
        # Karena data sampel sudah sorted, cek dahulu yang terakhir dari x dengan X, menghindari
        # mencari angka yang out of range,
        if X[len(X) - 1] < X[i]:
            print(f"{X[i]} out of range")
            continue
        # Ketika tidak out of range, kita akan loop sebanyak data x untuk mencari interval di mana
        # X berada di antara data sampel x
        else:
            for j in range(0, len(x)):
                # Karena sorted, ketika lebih besar tidak akan ada yang lebih besar lagi, maka di antara j.
                if(x[j] > X[i]):
                    # return i terhadap index interval
                    index = j - 1
                    # Stop loop
                    break
        # Hitung menggunakan rumus.
        hasilinterp = y[index] + ((y[index + 1] - y[index]) * (
            X[i] - x[index]) / (x[index + 1] - x[index]))
        # Append ke array yang tadi sudah dideklarasi.
        Y.append(hasilinterp)

    return Y
```

```python
# x atau data sample sudah ascending, sehingga tidak harus disortir kembali.
x = np.array([1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999])
y = np.array([14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577, 14.4187, 14.3438])
# Data yang akan diinterpolasi.
X = np.array([1982, 1984, 1986, 1988, 1990, 1992, 1994, 1996, 1998])

Y = my_lin_interp(x, y, X)

print("Hasil dari Linear Interpolasi")
# Print X dan Y
for i in range(len(X)):
    print(f"{X[i]} = {Y[i]}")

# Gambar Grafik
plt.figure(figsize=(12, 8))
plt.plot(x, y, "b")
plt.plot(x, y, "ro")
plt.plot(X, Y, "go")
plt.legend(["Garis Data Sample", "Titik Data Sample", "Linear Interpolasi"])
plt.show()
```
