```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
x = np.array([1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999])
X = np.array([1982, 1984, 1986, 1988, 1990, 1992, 1994, 1996, 1998])
y = np.array([
    14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577,
    14.4187, 14.3438
])

f = CubicSpline(x, y, bc_type="natural")
Y = f(X)
Y_rounded = []
print("Hasil dari Cubic Spline Interpolation")
for i in range(len(X)):
    print(f"{X[i]} = {round(Y[i],4)}")
    Y_rounded.append(round(Y[i],4))

plt.figure(figsize=(12, 8))
plt.title("Cubic Spline Interpolation")
plt.plot(x, y, "b")
plt.plot(x, y, "ro")
plt.plot(X, Y, "go")
plt.legend(["Garis Data Sample", "Titik Data Sample", "Linear Interpolasi"])
plt.show()
```
