```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
def divided_diff(x, y):
    n = len(y)
    # Buat array baru karena kita tidak ingin memodifikasi array sebelumnya.
    x = np.copy(x)
    a = np.copy(y)
    for k in range(1, n):
        a[k:n] = (a[k:n] - a[k - 1]) / (x[k:n] - x[k - 1])

    return a


def newton_poly(x, y, xtest):
    coef = divided_diff(x, y)
    n = len(x) - 1
    p = coef[n]
    for k in range(1, n + 1):
        p = coef[n - k] + (xtest - x[n - k]) * p
    return p
```

```python
x = np.array([1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999])
y = np.array([
    14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577,
    14.4187, 14.3438
])

X = np.array([1982, 1984, 1986, 1988, 1990, 1992, 1994, 1996, 1998])
Y = newton_poly(x, y, X)
Y_rounded = []

print("Hasil dari Quadratic Interplation (Newton polynom orde 2)")
for i in range(len(X)):
    print(f"{X[i]} = {round(Y[i],4)}")
    Y_rounded.append(round(Y[i], 4))

# Gambar Grafik
plt.figure(figsize=(12, 8))
plt.title("Interpolasi Quadratic (Newton Polynom Orde 2)")
plt.plot(x, y, "b")
plt.plot(x, y, "ro")
plt.plot(X, Y_rounded, "go")
plt.legend(["Garis Data Sample", "Titik Data Sample", "Linear Interpolasi"])
plt.show()
```
