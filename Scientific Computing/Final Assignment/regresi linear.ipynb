```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

# Deklarasi data x dan y.
x = np.arange(0, 540, 20)
y = np.array([
    0, 139, 298, 433, 685, 1026, 1279, 1373, 1490, 1634, 1800, 1986, 2191,
    2417, 2651, 2915, 3203, 3516, 3860, 4216, 4630, 5092, 5612, 6184, 6760,
    7327, 7581
])

A = np.vstack([x, np.ones(len(x))]).T
y = y[:, np.newaxis]

# Direct Least Squares Regression
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T, A)), A.T)), y)

# Anggap f(x) = ax + b
# Index pertama dari Direct Least Squares Regression adalah a
a = alpha[0]
# Index kedua dari Direct Least Squares Regression adalah b
b = alpha[1]
print(f"Didapatkan y = {a}x + {b}")
X = np.arange(0, 540, 20)
Y = a * X + b
print(f"{(a * X )+ b}")
```

```python
# Gambar Grafik
plt.figure(figsize=(12, 8))
plt.plot(x, y, "b")
plt.plot(x, y, "ro")
plt.plot(X, Y, "go")
plt.legend(["Garis Data Sample", "Titik Data Sample", "Linear Interpolasi"])
plt.show()
```

```python
# yNew = a*x + b
yNew = (a * x) + b
# Rubah size plot
plt.figure(figsize=(12, 8))
# Plot data awal
plt.plot(x, y, "b.")
# Plot data baru
plt.plot(x, yNew, "r")
plt.legend(["Data Awal", "Data dari Hasil Regresi Fungsi Linear"])
plt.title(f"y = {a}x + {b}")
plt.show()
```
