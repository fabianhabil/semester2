```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

# Deklarasi data x dan y.
x = np.array([1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999])
y = np.array([14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577, 14.4187, 14.3438])

A = np.vstack([x, np.ones(len(x))]).T
y = y[:, np.newaxis]

# Direct Least Squares Regression
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T, A)), A.T)), y)

# Anggap f(x) = ax + b
# Index pertama dari Direct Least Squares Regression adalah a
a = alpha[0]
# Index kedua dari Direct Least Squares Regression adalah b
b = alpha[1]
print(f"Didapatkan y = {a}x + {b}")
X = np.array([1982, 1984, 1986, 1988, 1990, 1992, 1994, 1996, 1998, 2000])
Y = np.array([14.2205, 14.13765, 14.151900000000001, 14.2333, 14.251249999999999, 14.2454, 14.2715, 14.3882, 14.38125, 14.351794444444444])
print(f"{(a * X )+ b}")
```

```python
# Gambar Grafik
plt.figure(figsize=(12, 8))
plt.plot(x, y, "b")
plt.plot(x, y, "ro")
plt.plot(X, Y, "go")
plt.legend(["Garis Data Sample", "Titik Data Sample", "Linear Interpolasi"])
plt.show()
```

```python
# yNew = a*x + b
yNew = (a * x) + b
# Rubah size plot
plt.figure(figsize=(12, 8))
# Plot data awal
plt.plot(x, y, "b.")
# Plot data baru
plt.plot(x, yNew, "r")
plt.legend(["Data Awal", "Data dari Hasil Regresi Fungsi Linear"])
plt.title(f"y = {a}x + {b}")
plt.show()
```
