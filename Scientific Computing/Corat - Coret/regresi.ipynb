```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
x = np.linspace(0, 10, 50)
y = 0.5 * np.exp(0.3 * x) + 0.5 * np.random.random(len(x))
```

```python
# plt.figure(figsize=(8,6))
plt.plot(x, y)
```

```python
# mencari nilai koefisien dari fungsi yang akan di-fit (Least Square Regression)
A = np.vstack([x, np.ones(len(x))]).T
b, log_a = np.linalg.lstsq(A, np.log(y), rcond=None)[0]
a = np.exp(log_a)
print(f"a={a}, b={b}")
```

```python
plt.plot(x, y, 'b')
plt.plot(x, a*np.exp(b*x), 'r')
plt.xlabel('x')
plt.ylabel('y')
# INGAT REGRESI GAK USAH NYENTUH TITIK, KALO NYENTUH TITIK ITU INTERPOLASI
```
