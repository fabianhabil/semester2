```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

```python
# Deklarasi data x dan y.
x = np.array([2.6, 3.1, 4.6, 6.5, 8.5, 11, 15.2, 34.7, 65.8, 69.4])
y = np.array([60, 120, 180, 240, 300, 360, 420, 480, 540, 600])

A = np.vstack([x, np.ones(len(x))]).T
y = y[:, np.newaxis]

# Direct Least Squares Regression
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T, A)), A.T)), y)

# Anggap f(x) = ax + b
# Index pertama dari Direct Least Squares Regression adalah a
a = alpha[0]
# Index kedua dari Direct Least Squares Regression adalah b
b = alpha[1]

print(f"Didapatkan y = {a}x + {b}")
```

```python
# Bikin data baru hasil dari regresi.
x_new = np.linspace(2.6, 69.4, 100)
y_new = a * x_new + b
# Cari y ketika x = 50
x_data = 50
y_data = a * x_data + b

plt.figure(figsize=(12, 8))
plt.xlabel("âˆ†x (cm)")
plt.ylabel("F (Newton)")
plt.title(f"y = {a}x + {b}")
plt.plot(x_new, y_new)
plt.plot(x, y, "ro")
plt.plot(x_data, y_data, "ko")
plt.legend(["Regresi Linear", "Titik Data", f"F saat 50cm = {y_data}"])
plt.show()
```

```python
# Import library pembantu CubicSpline
from scipy.interpolate import CubicSpline

# Deklarasi data x dan y.
x = np.array([11, 15.2, 34.7, 65.8, 69.4])
y = np.array([360, 420, 480, 540, 600])

# CubicSpline jadikan fungsi agar tinggal dipanggil.
f = CubicSpline(x, y)

# Bikin data baru hasil dari interpolasi.
x_new = np.linspace(11, 69.4, 100)
y_new = f(x_new)

# Cari y ketika x = 50
x_data = 50
y_data = f(x_data)
plt.figure(figsize=(12, 8))
plt.xlabel("âˆ†x (cm)")
plt.ylabel("F (Newton)")
plt.plot(x_new, y_new, "r")
plt.plot(x, y, "bo")
plt.plot(x_data, y_data, "ko")
plt.legend(["Cubic Spline Interpolation", "Titik Data", f"F saat 50cm = {y_data}"])
plt.show()
```
